**Linear Regression Model**

```python
# 1️⃣ Import libraries
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

# 2️⃣ Drop target and derived columns from features
# Make sure you remove 'price' and any column like 'price_crore'
X = df.drop(columns=['price', 'price_crore'], errors='ignore')  # errors='ignore' avoids crash if column doesn't exist
y = df['price']

# 3️⃣ Identify categorical columns
categorical_cols = X.select_dtypes(include='object').columns
print("Categorical columns:", categorical_cols)

# 4️⃣ Train-test split first
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42
)

# 5️⃣ One-hot encode categorical columns
X_train_encoded = pd.get_dummies(X_train, columns=categorical_cols, drop_first=True)
X_test_encoded = pd.get_dummies(X_test, columns=categorical_cols, drop_first=True)

# 6️⃣ Align test columns with train columns
X_test_encoded = X_test_encoded.reindex(columns=X_train_encoded.columns, fill_value=0)

# 7️⃣ Train Linear Regression
lr = LinearRegression()
lr.fit(X_train_encoded, y_train)

# 8️⃣ Predictions
y_pred = lr.predict(X_test_encoded)

# 9️⃣ Evaluation
mse = mean_squared_error(y_test, y_pred)
rmse = mse ** 0.5
r2 = r2_score(y_test, y_pred)

print("\nModel Evaluation:")
print("MSE:", mse)
print("RMSE:", rmse)
print("R2:", r2)

# 🔟 Optional: Feature coefficients
coeff_df = pd.DataFrame({
    'Feature': X_train_encoded.columns,
    'Coefficient': lr.coef_
}).sort_values(by='Coefficient', key=abs, ascending=False)

print("\nTop 10 Important Features:")
print(coeff_df.head(10))
```

R² = 0.865 is the “performance indicator” in regression. It means the model explains 86.5% of the variation in prices

**RandomForest Regressor**

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score

# Features and target
X = df.drop(columns=['price', 'price_crore'], errors='ignore')
y = df['price']

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# One-hot encode categorical columns
X_train = pd.get_dummies(X_train, drop_first=True)
X_test = pd.get_dummies(X_test, drop_first=True).reindex(columns=X_train.columns, fill_value=0)

# Train Random Forest
rf = RandomForestRegressor(n_estimators=200, random_state=42, n_jobs=-1)
rf.fit(X_train, y_train)

# Predict and evaluate
y_pred = rf.predict(X_test)
print("RMSE:", mean_squared_error(y_test, y_pred)**0.5)
print("R2:", r2_score(y_test, y_pred))

# Top 10 feature importances
print(pd.DataFrame({'Feature': X_train.columns, 'Importance': rf.feature_importances_})
      .sort_values('Importance', ascending=False).head(10))
```

**XGB00st Regressor**

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
from xgboost import XGBRegressor

# Features and target
X = df.drop(columns=['price', 'price_crore'], errors='ignore')
y = df['price']

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# One-hot encode categorical columns
X_train = pd.get_dummies(X_train, drop_first=True)
X_test = pd.get_dummies(X_test, drop_first=True).reindex(columns=X_train.columns, fill_value=0)

# Initialize XGBoost Regressor
xgb = XGBRegressor(
    n_estimators=200,     # number of trees
    learning_rate=0.1,    # step size shrinkage
    max_depth=6,          # max depth of each tree
    random_state=42,
    n_jobs=-1
)

# Train the model
xgb.fit(X_train, y_train)

# Predict and evaluate
y_pred = xgb.predict(X_test)
print("RMSE:", mean_squared_error(y_test, y_pred)**0.5)
print("R2:", r2_score(y_test, y_pred))

# Feature importance
feat_imp = pd.DataFrame({
    'Feature': X_train.columns,
    'Importance': xgb.feature_importances_
}).sort_values('Importance', ascending=False)

print("\nTop 10 Important Features:")
print(feat_imp.head(10))
```
